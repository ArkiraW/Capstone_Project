## Consumer Price Index

```python
cpi = pd.read_excel('../data/raw_data/cpi.xlsx')
```

```python
cpi.head()
```

```python
# Drop first column
cpi.drop(columns='หมวด',inplace=True)

# Reshape using melt
cpi = cpi.melt(id_vars=["ปี"], var_name="Month", value_name="cpi")

# Combine Year and Month into a single column
cpi["date"] = cpi["ปี"].astype(str) + " " + cpi["Month"]

# Drop unnecessary columns and reorder
cpi = cpi[["date", "cpi"]]
```

```python
cpi.head()
```

```python
# Convert Time column to datetime
cpi["date"] = pd.to_datetime(cpi['date'], format="%Y %B")

cpi.head()
```

## Consumer Price Index(Rice)

```python
cpi_rice = pd.read_excel('../data/raw_data/cpi(rice).xlsx')
```

```python
cpi_rice.head()
```

```python
# Drop first column
cpi_rice.drop(columns='หมวด',inplace=True)
```

```python
# Reshape using melt
cpi_rice = cpi_rice.melt(id_vars=["ปี"], var_name="Month", value_name="cpi_rice")

# Combine Year and Month into a single column
cpi_rice["date"] = cpi_rice["ปี"].astype(str) + " " + cpi_rice["Month"]

# Drop unnecessary columns and reorder
cpi_rice = cpi_rice[["date", "cpi_rice"]]
```

```python
# Convert Time column to datetime
cpi_rice["date"] = pd.to_datetime(cpi_rice['date'], format="%Y %B")

# Set date to index
cpi_rice.set_index("date", inplace=True)
```

```python
cpi_rice.head()
```
