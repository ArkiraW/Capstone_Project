## Water station

```python
# import the csv file
water_station = pd.read_csv("../data/raw_data/all_stations_data_full.csv")
```

```python
water_station.head()
```

```python
# Check the number of water station
water_station['station_id'].describe()
```

```python
water_station.columns
```

```python
# Drop unused columns
water_station.drop(columns=['station_name','รวม\n(ล้าน ลบ.ม.)', 'station_name', 'source_file'],inplace =True)
```

```python
# Rename headers
water_station.rename(columns={'พ.ศ.':'year','เม.ย.': 4,'พ.ค.':5,'มิ.ย.':6, 'ก.ค.':7, 'ส.ค.':8,'ก.ย.':9, 'ต.ค.':10,\
                              'พ.ย.':11, 'ธ.ค.':12,'ม.ค.':1, 'ก.พ.':2, 'มี.ค.':3,},inplace=True)
```

```python
# Drop unsed rows
water_station = water_station[water_station['year'] != 'ค่าเฉลี่ย']
```

```python
# Change B.E. to A.D.
water_station.loc[:, 'year']= water_station['year'].apply(lambda x: int(x)-543)
```

```python
water_station
```

```python
# Since this data uses Thai New Years, it is needed to assign the month 1,2,3 to be in the previous year
water_station_2 = water_station.filter(items=['year',1,2,3,'station_id'])
water_station_2.loc[:, 'year']= water_station_2['year'].apply(lambda x: int(x)+1)
water_station.drop(columns=[1,2,3],inplace=True)
```

```python
# concat water station with national years
water_station = water_station.merge(water_station_2,how='left',on=['year','station_id'])
```

```python
water_station.head()
```

```python
# Reshape data frame

# Melt the DataFrame
water_station = water_station.melt(id_vars=["year", "station_id"], var_name="month", value_name="value")
# Create a datetime column
water_station ["date"] = pd.to_datetime(dict(year=water_station["year"], month=water_station ["month"], day=1))
# Set the datetime as the index and sort
water_station  = water_station.set_index("date").sort_index()
# Drop unnecessary columns
water_station  = water_station.drop(columns=["year", "month"])
# Pivot the DataFrame
water_station = water_station.pivot(columns="station_id", values="value")
# Rename columns to add 'water_station_' prefix
water_station.columns = [f"water_station_{col.replace('.', '')}" for col in water_station.columns]
```

```python
# Change data type from object to float
water_station = water_station.replace(",", "", regex=True).astype(float)
```

# Concat all data

```python
# Concat all data
df = pd.concat([price,prod_index,price_index,export,inflation,cpi_rice,pol_rate,exr,\
                water_dam,fertilizer,water_station],axis=1)
# Rename index 
df.index.name = 'date'

# Set index to date time
df.index = pd.to_datetime(df.index)

# Sort index according to the date
df.sort_index(inplace=True)

# Select time-frame from Jan 2005 to Oct 2024
df = df.loc[pd.to_datetime("2006-01"):pd.to_datetime("2024-10")]
```
